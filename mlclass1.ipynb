```python
# You have to generate 1000 random strings of length between (2,10), Calculate their hash value and sort them according to it's hash value
# without using any special library
```

```python
import random 
alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
```

```python
Strings = []
for j in range(1000):
    Length = random.randint(2,10)
    Blkstr = ""
    for i in range(0,Length):
        Randno = random.randint(0,61)
        Blkstr += alphabets[Randno]
    Strings.append(Blkstr)
```

```python
#need amendments
#correct one is below

asciival = {}
for i in range(len(alphabets)):
    asciival[alphabets[i]]=65+i
print(asciival)
```

```python
# correct code
# make dictionary with ascii value

Asciival = {}
Asc = 65

for char in alphabets[0:26]:
    Asciival[char] = Asc
    Asc += 1
Asc = 97
for char in alphabets[26:52]:
    Asciival[char] = Asc
    Asc += 1
Asc = 48
for char in alphabets[52:62]:
    Asciival[char] = Asc
    Asc += 1
```

```python
print(Asciival)
```

```python
Blklist = []
for i in range(1000):
    Hashvalue = 0
    Primeno = 13
    Power = 1
    for char in Strings[i]:
        Hashvalue += Asciival[char] * Primeno**Power
        Power += 1
    Blklist.append(Hashvalue)
```

```python
D = {}
for char,Hashval in zip(Strings,Blklist):
    D[Hashval] = char
```
